Inicializa una lista vac ́ıa que contendr ́a las listas de valores de p ́ıxeles,
donde cada lista representar ́a un d ́ıgito, y una lista vac ́ıa que contendr ́a
las etiquetas (n ́umeros del 0 al 9) correspondientes a cada d ́ıgito en el
dataset.
Puedes usar:
* cols.pop(0): Elimina y devuelve el primer elemento de la lista cols. Este primer elemento es la etiqueta (n ́umero del d ́ıgito) correspondiente a la
imagen, que se almacena en la lista labels.
* labels.append(int(cols.pop(0))): Elimina y devuelve el primer elemento de cols, que es la etiqueta del digito, lo convierte en un entero y lo agrega a la lista labels.
* map(int, cols): Convierte todos los valores restantes en cols (que son cadenas de texto) a enteros.
* dataset.append(list(map(int, cols))): Convierte los valores restantes en cols (los p ́ıxeles de la imagen) a enterosy los agrega a dataset como una lista.

```python
dataset = []  # Lista para almacenar listas de valores de píxeles
labels = []   # Lista para almacenar etiquetas de los dígitos

# Recorrer todas las filas del dataset (excepto la primera si es encabezado)
for row in data:
    cols = row[:]  # Copia la fila para no modificar data directamente
    labels.append(int(cols.pop(0)))  # Extraer y guardar la etiqueta
    dataset.append(list(map(int, cols)))  # Convertir los píxeles y guardarlos en dataset

# Verificar los primeros elementos
print("Primeras 5 etiquetas:", labels[:5])
print("Primeros 5 píxeles de las imágenes:", dataset[:5])
```

## Ejercicio 5
Crea una funci ́on que, dado una posici ́on 1 ≤ k ≤ 10000, imprima el k-
 ́esimo d ́ıgito del conjunto de datos (es decir, la k- ́esima fila del archivo CSV) como una cuadr ́ıcula de 28 × 28 caracteres. M ́as espec ́ıficamente, debes mapear cada rango de valores de p ́ıxeles a los siguientes caracteres:
[0, 64) → " "
[64, 128) → "."
[128, 192) → "*"
[192, 256) → "#"

Por ejemplo, debes mapear la secuencia 0, 72, 192, 138, 250 a la cadena " .#*#". Para la Figura 1 (que es la imagen n ́umero 130 en el conjunto de datos), el resultado ser ́ıa lo que ves en la Figura 2.

Para ello necesitas mapear cualquier valor de p ́ıxel entre 0 y 255 al caracter que queremos mostrar.

Crea una funci ́on caracter utiliza un diccionario llamado rangos que mapea rangos de valores de p ́ıxeles (entre 0 y 255) a caracteres espec ́ıficos.

Estos caracteres son utilizados para representar diferentes niveles de intensidad de los p ́ıxeles, de modo que:

* Los valores de p ́ıxeles entre 0 y 63 se mapean al car ́acter " " (un espacio en blanco).
* Los valores de p ́ıxeles entre 64 y 127 se mapean al car ́acter ".".
* Los valores de p ́ıxeles entre 128 y 191 se mapean al car ́acter "*".
* Los valores de p ́ıxeles entre 192 y 255 se mapean al car ́acter "#".

La funci ́on recibe un valor de p ́ıxel como entrada. Luego, itera a trav ́es de los elementos del diccionario ranges, comparando el valor del p ́ıxel con los rangos definidos:

* Si el valor del p ́ıxel cae dentro de un rango particular, se devuelve el car ́acter asociado a ese rango.

Por ejemplo, si el valor del p ́ıxel es 70, caer ́a dentro del rango (64, 128), por lo que la funci ́on devolver ́a el car ́acter ".". Esto permite una rep-
resentaci ́on visual simplificada de im ́agenes en escala de grises mediante caracteres, donde la intensidad del p ́ıxel determina el car ́acter utilizado.

```python
def caracter(pixel):
    """Mapea un valor de píxel (0-255) a un carácter visual."""
    rangos = {
        (0, 64): " ",
        (64, 128): ".",
        (128, 192): "*",
        (192, 256): "#"
    }
    for (a, b), ch in rangos.items():
        if a <= pixel < b:
            return ch
    return " "  # En caso de error, devolver espacio en blanco
```

Una vez que tenemos esta funci ́on, podemos mapear cualquier vector de 784 dimensiones a la secuencia de caracteres correspondiente.
Con esto, puedes iterar sobre las filas y sobre las columnas para imprimir la cuadr ́ıcula resultante.
Es importante notar que, cada imagen en MNIST tiene un tama ̃no de 28 p ́ıxeles de ancho por 28 p ́ıxeles de alto, lo que da un total de 784 p ́ıxeles. Por lo tanto, para imprimir el car ́acter en la fila i y la columna j,
necesitaremos acceder al valor en la posici ́on 28 × i + j del vector de 784 dimensiones.

```python
def imprimir_digito(k, dataset):
    """
    Imprime el k-ésimo dígito del dataset como una cuadrícula de 28x28 caracteres.

    Parámetros:
        k (int): Índice de la imagen (1 ≤ k ≤ 10000).
        dataset (list): Lista de listas con valores de píxeles.
    """
    if not (1 <= k <= len(dataset)):
        print("Error: k fuera del rango válido.")
        return

    imagen = dataset[k - 1]  # La imagen está en la fila k-1 (índice base 0)

    for i in range(28):  # Itera sobre las filas de la imagen
        fila = "".join(caracter(imagen[28 * i + j]) for j in range(28))
        print(fila)  # Imprime la fila como una cadena

# Prueba con varias imagenes de números
for i in range(1, 10):
  imprimir_digito(i, dataset)
```

Inicializamos listas para almacenar los conteos de píxeles negros para '0' y '1'

```python
conteo_pixeles_cero = [0] * 784
conteo_pixeles_uno = [0] * 784
```

Iteramos sobre los datos y actualiza los conteos

```python
for i in range(len(labels)):
  if labels[i] == 0:
    for j in range(784):
      if dataset[i][j] >= 128:  # Umbral para considerar un píxel como negro
        conteo_pixeles_cero[j] += 1
  elif labels[i] == 1:
    for j in range(784):
      if dataset[i][j] >= 128:
        conteo_pixeles_uno[j] += 1
```

Calculamos las diferencias absolutas entre los conteos de píxeles

```python
diferencias = [abs(conteo_pixeles_cero[i] - conteo_pixeles_uno[i]) for i in range(784)]
```

Encuentramos el índice del valor máximo en la lista de diferencias

```python
diferencias = [abs(conteo_pixeles_cero[i] - conteo_pixeles_uno[i]) for i in range(784)]
```

Encuentramos el índice del valor máximo en la lista de diferencias

```python
indice_max_diferencia = diferencias.index(max(diferencias))
```

Convierte el índice lineal del píxel a coordenadas (fila, columna) de la cuadrícula 28x28

```python
fila = indice_max_diferencia // 28
columna = indice_max_diferencia % 28
```

```python
print(f"El píxel que mejor separa los dígitos '0' y '1' se encuentra en la fila {fila}, columna {columna}.")
print(f"Diferencia absoluta: {max(diferencias)}")
```
