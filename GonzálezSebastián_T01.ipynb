Calcula e imprime la media y la desviación estándar para cada una de las 4 columnas de medición (es decir, la longitud y el ancho del sépalo, la longitud y el ancho del pétalo).  

Recuerda que, para una lista dada de $n$ valores $x = (x_1, x_2, ..., x_n)$, la media $\mu$ y la desviación estándar $\sigma$ se definen respectivamente como:

$$
\mu = \frac{1}{n} \sum_{i=1}^{n} x_i
$$

$$
\sigma = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2}
$$

```python
# En una forma más limpia:
import csv
import math

def leer_columnas_csv(ruta_csv, num_columnas=4):
    """Lee un archivo CSV y devuelve una lista de listas con las columnas."""
    columnas = [[] for _ in range(num_columnas)]  # Crear listas vacías para cada columna

    with open(ruta_csv, newline='') as f:
        reader = csv.reader(f)

        for fila in reader:
            if len(fila) >= num_columnas:  # Verificar que la fila tenga suficientes columnas
                for i in range(num_columnas):
                    columnas[i].append(float(fila[i]))  # Convertir a float

    return columnas

def calcular_suma(lista):
    """Calcula la suma de una lista."""
    return sum(lista)

def calcular_media(lista):
    """Calcula la media de una lista."""
    return sum(lista) / len(lista) if lista else 0  # Evitar división por cero

def calcular_desviacion_estandar(lista, media):
    """Calcula la desviación estándar de una lista."""
    n = len(lista)
    return math.sqrt(sum((x - media) ** 2 for x in lista) / n) if n > 0 else 0

# Ruta del archivo CSV
ruta_csv = "/content/drive/MyDrive/TSFC3_Ciencia_de_Datos/iris.data"

# Leer columnas
columnas = leer_columnas_csv(ruta_csv)

# Calcular e imprimir resultados
for i, lista in enumerate(columnas, start=1):
    suma = calcular_suma(lista)
    media = calcular_media(lista)
    desviacion = calcular_desviacion_estandar(lista, media)

    print(f"La suma de la lista {i} es: {suma}")
    print(f"La cantidad de elementos en la lista {i} es: {len(lista)}")
    print(f"La media de la lista {i} es: {media}")
    print(f"La desviación estándar de la lista {i} es: {desviacion}\n")
```

## 3.Imprime la media y la desviación estándar por separado
Calcula e imprime la media y la desviación estándar para cada una de las 4 columnas de medici ́on, por separado para cada una de las tres especies de Iris (versicolor, virginica y setosa).

```python
import csv
import math
from collections import defaultdict

def leer_datos_csv(ruta_csv):
    """Lee un archivo CSV y organiza los datos en un diccionario por especie."""
    datos_por_especie = defaultdict(lambda: [[] for _ in range(4)])  # Diccionario de listas vacías para cada especie

    with open(ruta_csv, newline='') as f:
        reader = csv.reader(f)

        for fila in reader:
            if len(fila) >= 5:  # Verificar que la fila tenga al menos 5 columnas
                especie = fila[4].strip()  # Obtener la especie y eliminar espacios extra
                for i in range(4):
                    datos_por_especie[especie][i].append(float(fila[i]))  # Agregar valores numéricos

    return datos_por_especie

def calcular_media(lista):
    """Calcula la media de una lista."""
    return sum(lista) / len(lista) if lista else 0

def calcular_desviacion_estandar(lista, media):
    """Calcula la desviación estándar de una lista."""
    n = len(lista)
    return math.sqrt(sum((x - media) ** 2 for x in lista) / n) if n > 0 else 0

# Ruta del archivo CSV
ruta_csv = "/content/drive/MyDrive/TSFC3_Ciencia_de_Datos/iris.data"

# Leer datos organizados por especie
datos_por_especie = leer_datos_csv(ruta_csv)

# Procesar cada especie y calcular estadísticas
for especie, columnas in datos_por_especie.items():
    print(f"\nEstadísticas para {especie}:")
    for i, lista in enumerate(columnas, start=1):
        media = calcular_media(lista)
        desviacion = calcular_desviacion_estandar(lista, media)

        print(f"  Columna {i}: Media = {media:.2f}, Desviación Estándar = {desviacion:.2f}")
```

## 6. Ajustar una curva de densidad de probabilidad normal
Una vez que tengas el histograma, puedes ajustar una curva de densidad de probabilidad normal sobre  ́el. Para crear esta curva primero necesitas calcular la media (mean) y la desviaci ́on est ́andar (std) de los valores de cada especie para la caracteristica en cuestion.
Luego, genera una serie de valores x que cubran un rango alrededor de la media:
x = np.linspace(u - 5*s, u + 5*s, 100)
• u: Media de los valores para la caracter ́ıstica y especie.
* s: Desviaci ́on est ́andar de los valores.
* np.linspace(u - 5*s, u + 5*s, 100): Genera 100 puntos x entre u - 5*s y u + 5*s para cubrir un rango amplio alrededor de la
media.

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
from scipy.stats import norm  # Justo para la curva de densidad normal

# Repito lo mismo del de arriba
for especie, columnas in datos_por_especie.items():
    print(f"\n Listas generadas para {especie}:")

    plt.figure(figsize=(10, 6))

    for i, lista in enumerate(columnas):
        print(f"  Lista {i+1} ({nombres_columnas[i]} - {especie}): {lista[:5]}...")  # Muestra los primeros 5 valores

        # Acá agregamos esto para calcular media y desviación estándar
        u = np.mean(lista)  # Media
        s = np.std(lista)   # Desviación estándar

        # Ahora generemos los valores x para la curva de densidad normal
        x = np.linspace(u - 5*s, u + 5*s, 100)
        y = norm.pdf(x, u, s)  # Función de densidad de probabilidad

        # Graficar histograma y superponer la curva de densidad
        plt.hist(lista, density=True, alpha=0.5, color=colores_columnas[i], bins=15, label=nombres_columnas[i])
        plt.plot(x, y, color=colores_columnas[i], linestyle="--", linewidth=2)

    plt.title(f"Histogramas y Curvas Normales de {especie}")
    plt.xlabel("Valor")
    plt.ylabel("Densidad")
    plt.legend()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
from scipy.stats import norm

for i in range(4):
    plt.figure(figsize=(10, 6))  # Crear una nueva figura en cada iteración

    for especie in especies:
        if especie in datos_por_especie:
            lista = datos_por_especie[especie][i]  # Obtener los datos de la característica
            color = colores_especies[especie]

            # Crear histograma de densidad
            plt.hist(lista, bins=15, density=True, alpha=0.5, color=color, label=especie)

            # Calcular y dibujar la curva de densidad normal
            u = np.mean(lista)  # Media
            s = np.std(lista)   # Desviación estándar
            x = np.linspace(u - 5*s, u + 5*s, 100)
            y = norm.pdf(x, u, s)
            plt.plot(x, y, color=color, linestyle="--", linewidth=2)

    plt.title(f"Distribución de {nombres_columnas[i]}")
    plt.xlabel("Valor")
    plt.ylabel("Densidad")
    plt.legend()
    plt.tight_layout()
    plt.show()
```
